```python
#Problem 1: Unique Words in a Sentence
sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words) 

print("Unique words count:", len(unique_words)) #Unique word count
print("Unique words:", unique_words)
```

```python
#Problem 2: Highest Salary from Employee Data

#Using Dictionary
employees = {"Ravi": 75000, "Anita": 68000, "Kiran": 72000}

highest_employee = ""
highest_salary = 0

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#Problem 4: Count Products Above a Price Threshold

numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

#for loop through each number
for num in numbers: 
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
#Problem 5: Calculate Attendance Percentage

#P- Present and A - Absent
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100
print("Attendance Percentage:", percentage)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone = [9876543210, 9123456789, 9876543210]

#list to set conversion
unique = set(phone)

print("Unique phone numbers:", unique)
```

```python
#Problem 7: Count Character Frequency
text = "python"

# empty dictionary to store character counts
char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1 #Count increment 
    else:
        char_count[char] = 1

print(char_count)
```

```python
#Problem 10: Calculate Average Marks
marks = [65, 75, 80, 70]

total = 0

for mark in marks:
    total += mark

avg = total / len(marks)

print("Average Marks:", avg)
```
