```python
## Task 1: User Login Check
# Define correct credentials
correct_username = "admin"
correct_password = "1234"

# Given user input 
username = "admin"
password = "1234"

# Check if both username and password match
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
## Task 2: Pass / Fail Analyzer
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Loop through each mark
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final results
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

```python
## Task 3: Simple Data Cleaner
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean each name
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
## Task 4: Message Length Analyzer
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)
    
    if length > 10:
        print("This message is longer than 10 characters.")
    
    print()  
```

```python
## Task 5: Error Message Detector
# System log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count number of ERROR entries
for log in logs:
    if log == "ERROR":
        error_count += 1

# Print total error count
print("Total ERROR entries:", error_count)
```
