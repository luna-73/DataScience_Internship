## Task 1 — Conceptual Understanding

### Q1. What is the difference between "Love" and "love" in NLP?

In NLP, text is treated as case-sensitive by default, which means "Love" and "love" are considered two different tokens even if the meaning of both the words is same. This can lead to duplication in features and inconsistent analysis. To avoid this, we usually convert all text to lowercase during preprocessing so that both forms are treated as a single, consistent token.

---

### Q2. What happens if stopwords are not removed?

Stopwords are very common words such as "the", "is", and "and" that usually do not carry significant meaning. If they are not removed, they tend to dominate frequency counts and introduce noise into the data. This can make it harder for models to focus on meaningful words and can also increase computational overhead unnecessarily.

---

### Q3. Provide two real-world scenarios where removing stopwords can be harmful.

1. **Sentiment Analysis:**  
   In a sentence like "I am not happy with this product", the word "not" plays a critical role in expressing negative sentiment. If stopwords are blindly removed, the sentence may become "I happy with this product", which completely flips the meaning. This can lead to incorrect sentiment predictions and unreliable model outputs.

2. **Search or Question-Based Systems:**  
   In queries such as "Who is the CEO of Tesla?", words like "who" and "is" help define the intent of the question. Removing them may reduce the query to just keywords like "CEO Tesla", which can lose the original context and return less accurate or irrelevant results.
---

### Q4. Explain the difference between stemming and lemmatization.

Both stemming and lemmatization aim to reduce words to their base form, but they do it differently. Stemming applies simple rule-based trimming, which can sometimes produce incomplete or non-meaningful words (e.g., "running" → "runn"). Lemmatization, on the other hand, uses vocabulary and linguistic knowledge to return the correct base form (e.g., "running" → "run"), making it more accurate and meaningful.

# Task 2: Building the main preprocessing function


```python
def preprocess_text(txt):
    if not txt or not isinstance(txt, str):
        return [], ""

    # lower
    txt = txt.lower()

    # remove links & emails
    txt = re.sub(r'http\S+|www\S+|https\S+', '', txt)
    txt = re.sub(r'\S+@\S+', '', txt)

    # remove digits
    txt = re.sub(r'\d+', '', txt)

    # remove repeated chars
    txt = re.sub(r'(.)\1{2,}', r'\1', txt)

    # remove emojis
    txt = emoji.replace_emoji(txt, replace='')

    # remove symbols
    txt = re.sub(r'[^a-z\s]', '', txt)

    # clean spaces
    txt = re.sub(r'\s+', ' ', txt).strip()

    tok = txt.split()

    out = []
    for w in tok:
        if len(w) > 2 or w in ['no', 'not']:
            out.append(w)

    sent = " ".join(out)

    return out, sent
```

# Task 3: Creating diverse test cases to evaluate preprocessing


```python
txts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

# Task 3: Applying preprocessing on all test sentences
Displaying original text, tokens, and cleaned output

```python
res = []

for t in txts:
    tok, sent = preprocess_text(t)

    print("\nOriginal:", t)
    print("Tokens:", tok)
    print("Cleaned:", sent)

    res.append((tok, sent))
```

# Task 4: Performing token-level analysis
Calculating total tokens, unique tokens, and average token length

```python
for i, (tok, sent) in enumerate(res):
    avg = sum(len(w) for w in tok) / len(tok) if tok else 0

    print(f"\nSentence {i+1}")
    print("Total tokens:", len(tok))
    print("Unique tokens:", len(set(tok)))
    print("Avg token length:", round(avg, 2))
```

## Analysis

# Task 5: Frequency analysis using Counter
Finding most frequent and least frequent words

```python
all_tok = []

for tok, _ in res:
    all_tok.extend(tok)

freq = Counter(all_tok)

print("Top 10 frequent words:")
print(freq.most_common(10))

print("\nLeast frequent words:")
print(freq.most_common()[-5:])
```

# Task 6: Building full pipeline function
Processes a list of texts and returns combined tokens and cleaned sentences

```python
def full_pipeline(txt_list):
    all_tok = []
    clean_sents = []

    for t in txt_list:
        tok, sent = preprocess_text(t)
        all_tok.extend(tok)
        clean_sents.append(sent)

    return {
        "tokens": all_tok,
        "clean_sentences": clean_sents
    }
```

# Task 6: Running the full pipeline on input data

```python
out = full_pipeline(txts)
print(out)
```

# Task 7: Testing edge cases (empty string, emojis, numbers)
Ensuring function handles invalid or unusual inputs gracefully

```python
tests = ["", "😍😍😍", "123456"]

for t in tests:
    print("\nInput:", t)
    print(preprocess_text(t))
```
