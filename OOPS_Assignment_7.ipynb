```python
#Problem 1: Smart Light Controller
class SmartLight:
    
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default status
    
    def switch(self, action):
        if action.upper() == "ON":
            self.status = "ON"
        elif action.upper() == "OFF":
            self.status = "OFF"
    
    def show_status(self):
        print(f"{self.name} is {self.status}")


light = SmartLight("Bedroom Light")
light.switch("ON")
light.show_status()
```

```python
#Problem 2: Employee ID Card System
class Employee:
    
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept
    
    def display_id(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


emp = Employee("Rahul", "EMP102", "IT")
emp.display_id()
```

```python
#Problem 3: Mobile Contact Record
class Contact:
    
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c1 = Contact("Anita", 9876543210)
c1.show_contact()
```

```python
#Problem 4: Product Price Tag Generator
class Product:
    
    def __init__(self, pname, price):
        self.pname = pname
        self.price = price
    
    def show_tag(self):
        print("Product:", self.pname)
        print(f"Price: ₹{self.price}")


p1 = Product("Headphones", 2499)
p1.show_tag()
```

```python
#Problem 5: Movie Rating Display System

class Movie:
    
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    def show_movie(self):
        print("Movie:", self.name)
        print(f"Rating: {self.rating} / 5")


m1 = Movie("Inception", 4.8)
m1.show_movie()
```
