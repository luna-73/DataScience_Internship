```python
# Problem 1: Post Engagement Analyzer
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 300, 250, 300]
analyze_engagement(likes)
```

```python
#Problem 2: Medicine Stock Alert System
def check_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print("Medicine Stock:", stock)
    print("Status:", status)

check_stock(6)
```

```python
#Problem 3: Rainfall Adequacy Checker
def check_rainfall(rainfall_list, required_level):
    total = 0
    
    # Calculate total rainfall
    for rain in rainfall_list:
        total += rain
    
    average = total / len(rainfall_list)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rain_data = [60, 75, 80, 70]
check_rainfall(rain_data, 70)
```

```python
#Problem 4: Duplicate Account Detection
def detect_duplicates(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["user1", "user2", "user1", "user3"]
detect_duplicates(users)
```

```python
#Problem 5: Appointment Eligibility Checker
def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_eligibility(21)
```

```python
#Problem 6: Premium Crop Price Filter
def filter_premium_crops(prices):
    premium = []
    
    # Filter prices above 2000
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)

crop_prices = [1500, 2500, 3200, 1800]
filter_premium_crops(crop_prices)
```

```python
#Problem 7: Application Health Checker
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)

check_system_health(7)
```

```python
Problem 8: Daily Transaction Limit Checker
def check_transaction(amount):
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

check_transaction(60000)
```
